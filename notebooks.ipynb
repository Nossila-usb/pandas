### Pandas

- Pandas é uma Biblioteca do Python voltada para a estruturação de dados;
- Permite a leitura e escrita de dados em vários formatos: **CSVs**, **Excel**, **SQL**, entre outros;
- Permite a realização de análise de dados e modelagem;
